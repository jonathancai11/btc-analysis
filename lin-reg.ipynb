## Linear Regression
  
Just testing out sklearn.linear_model.LinearRegression 

1. Generate random data set  
2. Visualize  
3. Find line of best fit  
4. Change variation and see accuracy?

```python
import numpy as np

from sklearn import svm, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
size = 1000
mean = 30
sd = 5
y = np.random.normal(mean, sd, size)
x = np.arange(size)
y += x / 40
```

```python
plt.scatter(x, y)
```

```python
percent_training = .8
X_train, X_test, y_train, y_test = train_test_split(x.reshape(-1, 1), y.reshape(-1, 1), test_size=(1 - percent_training))
display(X_train.shape)
display(X_test.shape)
```

```python
clf = LinearRegression(n_jobs=-1)
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
accuracy
```

```python
xs_predictions = np.arange(1500)
predictions = clf.predict(xs_predictions.reshape(-1, 1))
plt.scatter(x, y)
plt.scatter(xs_predictions, predictions)
```
